# CCTV 데이터와 인구 데이터 합치고 분석하기
- '구별' 공통컬럼을 기준으로 데이타 합치기
- 데이타프레임명 = pd.merge(데이타프레임1, 데이타프레임2, on=공통컬럼명)

````python
data_result = pd.merge(CCTV_Seoul, pop_seoul, on='구별')
# data_result.head(2)
````

### 2013년도 이전,2014년,2015년, 2016년 컬럼 삭제하기

````python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()
````

# 특정컬럼을 인덱스로 변경하기

### mathplotlib로 막대그래프 그리기

````python
# cctv 수치를 기준으로 그래프 그리기 
plt.figure()
data_result['소계'].plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
````

````python
# cctv 수치를 역순정렬후 그래프 그리기 
data_result['소계'].sort_values().plot(kind='barh', 
                                     grid=True, figsize=(10,10))
plt.show()
````

### mathplotlib로 산점도 그래프 그리기

````python
# 인구수와 CCTV 갯수의 산점도 그리기 
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
````

### 산점도+ 회귀선 그래프 추가하기

````python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
fp1
````

````python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
````

````python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
````

### sns의 regplot() 그래프로 산점도+회귀선 그래프 그리기

````python
ax = sns.regplot(x='인구수', y='소계', data=data_result)
ax
````
