### displot()을 이용한 히스토그램 그리기
- 1차원만 가능
- 데이터프레임[컬럼명] => 시리즈화

## 히스토그램 + 밀집도선 + 러그 그래프
- sns,distplot(데이터프레임[컬럼명], rug=True)
- rug : 양탄자 그래프는 그래프의 축에 동일한 길이의 직선을 붙여 밀집 정도를 표현한 그래프

```python
# 기본틀과 격자 구성
# fig = plt.figure()
# ax = plt.subplots(1,1,1)
ax = plt.subplots()
# 그래프로 데이터 전달 후 제목, 라벨 설정

# 히스토그램과 밀집도 displot에 양탄자(Rug) 그래프 추가
# 양탄자 그래프는 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프
ax = sns.distplot(tips['total_bill'], rug=True)

ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
ax
```

## count 그래프 : 특정 데이터값의 이산값을 나타낸 그래프
- 시리즈 형태
- sns.countplot('컬럼명', data=데이터프레임명)

```python
# 'day' 컬럼의 중복 없애기
tips['day'].unique() # 함 봣어요그냥 ( 어떤 값들로 이루어져 있는지)

# 데이터값의 빈도수 확인
tips['day'].value_counts()
```

## 산점도 그래프
- fit_reg=True/False
- sns.regplot(x='컬렴명1', y='컬럼명2', 'data=데이터프레임명, fit_req=True/False)

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
# ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatter of Total bill and Tip')  # 제목 지정
ax.set_xlabel('Total Bill') # X축 제목
ax.set_ylabel('tip') # Y축 제목
```

### lmplot()을 이용하여 산점도에 색상 추가하기

- 카테고리 데이터형에 이용
- 색상을 이용해서 산점도 표시
- sns.lmplot(x'컬럼명1(수치)', y='컬럼명2(카테고리형)', data=데이터프레임이름, hue='컬럼명2(카테고리형)', fit_reg=False/True)

```python
# 성별+ 총 팁 관의 상관관계
ax = sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False)
ax
```

## 2차원 밀집도 그래프(kdeplot)

- shade=True/False : 그래프의 음영효과 여부 
- sns.kdeplot(dataFrame['컬럼명1'], data2=dataFrame['컬럼명2'], shade=True/False)
    

```python
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], data2=tips['tip'], shade=True, color='green')

ax.set_title('Kernel Density of Total bill and Tip')
ax.set_xlabel('Total BIll')
ax.set_ylabel('tip')
```

### jointplot을 이용한 산점도 + 히스토그램

- kind='hex'사용시 육각형 그래프로 표시
- 육각 그래프란? 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프
- 데이터의 개수가 많아지면 점점 진한색으로 표현된다.
- sns.joinplot(x='컬럼명1', y='컬럼명2', data=데이터프레임명, kind='hex')
- kind = { 'scatter' | 'reg' | 'resid' | 'kde' | 'hex'}

```python
# jointplot에 육각그래프(hexbin)이용하기 : kind='hex'
ax = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
ax.fig.suptitle('Joint Plot', fontsize=10, y=1.03)
ax.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
ax
```

### 바이올린 그래프(violin plot)

- 바이올린 그래프란? 데이터의 분산을 확실하게 하기 위해 커널밀도를 추정한 그래프
- 박스플랏의 박스 형태를 바이올린 형태로 변경한 그래프
- sns.violinplot(x='컬럼명1', y='컬럼명2', data=tips)
