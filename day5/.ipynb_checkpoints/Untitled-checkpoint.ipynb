# Bar Plot 1 : 세로막대그래프 ( plt.bar 방식)
- plt.bar(xData, yData, label=labelName, color=colorName, width=float)
- width : 막대가로크기. 1보다 작게 지정

```python
# 데이터 생성 
x = list(range(10))
y = np.random.randint(10,100,10)

# 피규어 생성 
fig14 = plt.figure(dpi=200)

# plt.bar()를 이용하여 데이타 연결 => 세로막대 그래프 
fig14, plt.bar(x,y,width=0.85, color='red')

# 제목, 그리드, 눈금, 축 속성 지정 
fig14, plt.title('Bar Plot Sample')
fig14, plt.xlim(-1, 10)
fig14, plt.xticks(list(range(0,11)))
fig14, plt.yticks(list(range(10,110,10)))

plt.show()
```

# Bar Plot 2 : 세로막대그래프 ( add_subplot 방식)
- fig 객체 생성
    fig객체변수 = plt.figure(옵션)

- add_subplot()를 이용하여 그래프 객체 생성
    axes변수 = fig.add_subplot(행수, 열수, 순서)

- x, y 데이타 생성

- 막대 그래프 그리기

    axes.bar(xData, yData, label=labelName, color=colorName, width=float)

- X축과 Y축 설정하기
    axes.set_xticks([i for i in range(0,50,5)])

    axes.set_yticks([i for i in range(0,50, 5)])

    axes.set_xlabel('X 축')

    axes.set_ylabel('Y 축')

    axes.set_xlim([0, 50])

    axes.set_ylim([0, 50])

- 그리드 보이기
    axes.grid(color='g', linestyle='--', linewidth=1, alpha=0.5)

- 범례 보이기
    axes.legend()

- 그래프 제목 지정하기
    axes.set_title(그래프제목)

- 그래프 그리기
    plt.show()

```python
# 1) fig 변수 생성 
fig = plt.figure(dpi=150)
# 2) ax변수 = fig변수.add_subplot(행,열,순서)
axes = fig.add_subplot(1,1,1)
# 3) 데이타 생성 
x = range(1,11)
y = np.random.randint(10,100,10)
# 4) 데이타 연결 => 그래프 생성 
# ax변수.bar(xData, yData, width=가로넓이, color=막대그래프색상 )
axes.bar(x,y,width=0.9,color='red')
# 5)제목 지정 
# ax변수.set_title(그래프제목)
axes.set_title('막대그래프 예제')
# - X축과 Y축 설정하기 
axes.set_yticks([i for i in range(0,100,10)])
axes.set_xlabel('X  축')
axes.set_ylabel('Y  축')
axes.set_xlim([0, 11])
axes.set_ylim([0, 100])
plt.show()
```
