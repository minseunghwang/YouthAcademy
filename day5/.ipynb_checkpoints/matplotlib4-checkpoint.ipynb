# Mathplotlib에서 그래프를 그리는 2가지 방법

- plt.plot()
- add_subplot()

# 그리드 속성
```
plt.grid(linestyle='--', # 구분선모양
             color='g',     # 선의 색깔
             linewidth='1', # 굵기
             alpha=0.5)     # 투명도
```

````python
# plt.plot() 방식으로 다중 그래프 그리기

x = np.arange(0,12,0.1)
y = np.sin(x)

fig = plt.figure(dpi=150)
plt.plot(x,y, label='sin1', c='m', linestyle='--')
plt.plot(x,-y, label='sin2', linestyle='-.')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('다중 그래프')
plt.legend()
plt.grid(linestyle='--', color='g', linewidth=1,alpha=0.5)

# 축 속성 설정하기
# 축의 최소/최대 범위 설정
plt.xlim([0,15])
plt.ylim([-1.5,1.5])

# 축의 눈금 설정
x_ax = list(range(0,16,3))
y_ax = list(np.arange(-1.5,1.5,0.3))
plt.xticks(x_ax)
plt.yticks(y_ax)


plt.show()
````

````python
# add_subplot() 방식으로 다중 그래프 그리기

# 1) fig 객체 생성
fig = plt.figure(dpi=120)
# plt.show()

# 2) ax 변수 생성
# ax변수 = fig객체변수.add_subplot(행수,열수,순서)
ax1 = fig.add_subplot(1,1,1)
# ax

# 3) 데이터 생성
# np.linspace(start, end, size)
x = np.arange(0,12,0.1)
y = np.sin(x)

# 4) 데이터 전달해서 그래프 생성
# 1번 그래프
ax1.plot(x,y,label='sin1', c='m', linestyle='--')
# 2번 그래프
ax1.plot(x,-y, label='sin2', linestyle='-.')

# 5) 제목 표시
ax1.set_title('다중 그래프')


# 6) 그리드 표시
ax1.grid(linestyle='--', color='g', linewidth='1', alpha=0.5)


# 7) 범례 표시
ax1.legend(loc=1, facecolor='lightblue', fontsize=15)

# 8) 축 제목 표시
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')


# 9) 축의 눈금 변경
# 축 속성 설정
ax1.set_xlim([0,15])
ax1.set_ylim([-1.5,1.5])

# 축의 눈금 설정하기
x_ax = list(range(0,16))
ax1.set_xticks(x_ax)

y_ax = np.arange(-1.5,1.5)
ax1.set_yticks(y_ax)

plt.show()
````

````python
# plot() 방식으로 시리즈 그래프 그리기 
s1 = pd.Series(np.random.randn(10))
# s1 # -2.0 ~ 2.0
````

````python
fig = plt.figure(dpi=140)
s1.plot(c='r', linestyle='-.', marker='X', markerfacecolor='hotpink',  
                      markersize=10, label='S1')
plt.legend()
plt.grid(linestyle='--', color='g', linewidth=1, alpha=0.5)

# 축 속성 설정하기 
plt.xlim([-1,10])
plt.ylim([-2.0, 2.0])

# 축의 눈금 설정하기 
x_ax = list(range(0,12))
plt.xticks(x_ax)
y_ax = np.arange(-2.0, 2.0, 0.4)
plt.yticks(y_ax) 

plt.show()
````

````python
s1 = pd.Series(np.random.randn(10))
````

````python
# 1) fig 객체 생성
fig2 = plt.figure(dpi=120)
# 2) ax 변수 생성 
# ax변수 = fig객체변수.add_subplot(행수,열수,순서)
ax2 = fig2.add_subplot(1,1,1)
# 3) 데이타전달 후 그래프 생성 
ax2.plot(s1, c='r', linestyle='-.', marker='X', markerfacecolor='green',  
                      markersize=10, label='S1')
# 4) 그리드/축/범례/눈금 설정
ax2.legend()
ax2.grid(linestyle='--', color='g', linewidth=1, alpha=0.5)
# 축 속성 설정하기 
ax2.set_xlim([-1,10])
ax2.set_ylim([-2.0, 2.0])
# 축의 눈금 설정하기 
x_ax = list(range(0,12))
ax2.set_xticks(x_ax)
y_ax = np.arange(-3.0, 2.8, 0.4)
ax2.set_yticks(y_ax) 
plt.show()
````
