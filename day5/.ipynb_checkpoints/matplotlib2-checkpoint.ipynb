````python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import os
````

# 시리즈로 다중 그래프 만들기

- 1) 시리즈 데이터 생성하기
- 2) fig 객체 생성
    - fig변수 : plt.figure(옵션속성=값)
    - figsize는 ~10. 그래프 전체틀 크기
    - facecolor = 영문 색상명이나 16진수 색상값 # 배경 색상
    - frameon = True/False # 테두리 프레임 Show/Hide
    - degecolor = 영문 색상명이나 16진수 색상값 # 테두리 색상
    - dpi 해상도 : 숫자가 높을수록 해상도와 크기가 증가

- 3) 데이터 전달
    ```
    시리즈변수1.plot(옵션속성=값)
    시리즈변수2.plot(옵션속성=값) 
    ```
- 4) 제목 지정
```
plt.title('그래프제목, fontdict=[
```

````python
# 데이터 생성
x = np.arange(0,40)
# print(f'x : {x}')

print('*'*50)
# 시리즈 생성
s1 = pd.Series(np.random.randint(40, 80, size=40))
# print(s1)
````

````python
# figure 객체 생성
fig = plt.figure(figsize=[12,7], facecolor='AntiqueWhite', frameon=True, edgecolor='#0088ff', dpi=60, linewidth=20)

# 시리즈 데이터를 그래프에 전달
# 시리즈변수.plot()
s1.plot(marker='+', markerfacecolor='green', markersize=7.5)

# linestyle ; solid, dashed, dotted, dash_dot

# 2번째 시리즈 그래프 생성
s2 = pd.Series(np.random.randint(0,80,size=40))
s2.plot(linestyle='dashed', color='hotpink')

# 3번째 시리즈 그래프 생성
s2 = pd.Series(np.sin(x))
s2.plot(linestyle='solid', color='blue')

# 그래프 제목 지정
# 글꼴 : Geoigia / Arial /
# C:\Windows\Fonts
plt.title('MultiGraph', pad=50, loc='right',fontdict={'fontsize':50, 'color':'#00d5ff', 'fontweight':'bold', 'fontfamily':'Broadway'})


# loc는 1 2 3 4 있는데 1사분면, 2사분면 인가보다 (? 근데 10까지 있네)
plt.legend(labels=['S1','S2','S3'], facecolor='yellowgreen', loc=1, labelspacing=3, edgecolor='crimson',)

plt.show()
````

# 한글 폰트 사용하기

plt.rcParams['font.family']

### 글꼴 변경하기

````python
# 현재 설정되있는 폰트 확인
print(plt.rcParams['font.size'])
print(plt.rcParams['font.family'])
````

````python
# malgun으로 글꼴을 바꿔보겠다.
plt.rcParams['font.family'] = 'Malgun Gothic'
print(plt.rcParams['font.family'])

# 바뀐것 check
````

````python
# 한글 확인하기
fig = plt.figure(dpi=100)
s4 = pd.Series(np.random.randint(0,80, size=40))
s4.plot(linestyle='dashed', color='hotpink')
plt.title('시리즈 그래프')
plt.show()
````
