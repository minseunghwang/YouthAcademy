## subplot 이용하기 3 - add_subplot
- 피규어변수 = plt.figure()
- 변수 = 피규어변수.add_subplot(행,열,순서)
- plt.tight_layout() # 작게 레이아웃 배치
- 피규어변수, 변수.plot(x데이타, y데이타, 옵션)

````python
# 2*2 => 4개 

# 데이타 생성 
x =  ['a','b','c','d']
y1 = [0, 5, 10, 15]
y2 = [10, 5, 10, 15]
y3=  [20, 5, -10, 15]
y4 = [0, 5, 10, 15]

fig8 = plt.figure()

# 서브플랏 변수 생성 
ax1 = fig8.add_subplot(2,2,1)
ax2 = fig8.add_subplot(2,2,2)
ax3 = fig8.add_subplot(2,2,3)
ax4 = fig8.add_subplot(2,2,4)

# 데이타 연결
ax1.plot(x, y1, label='ax1')
ax2.plot(x, y2, label='ax2')
ax3.plot(x, y3, label='ax3')
ax4.plot(x, y4, label='ax4')

# 그리드 표시 
ax1.grid()
ax2.grid()
ax3.grid()
ax4.grid()

# 각각 그래프의 제목
ax1.set_title('그래프1')
ax2.set_title('그래프2')
ax3.set_title('그래프3')
ax4.set_title('그래프4')

# 전체 그래프 제목
# fig8.suptitle('2*2 그래프', x=0.9, y=1.1)
plt.suptitle('2*2 그래프', x=0.9, y=1.1)

# 범례표시 
fig8.legend(loc='lower center', facecolor='#cccccc')

plt.tight_layout() 
plt.show()
````

# 알고나풀어볼까

````python
# 짝지어 제거하기
from collections import deque
def solution(s):
    s=deque(list(s))
    stack = []
    
    if len(s)%2 != 0:
        return 0
    
    while s:
        if len(stack) == 0:
            stack.append(s.popleft())
        elif s[0] == stack[-1]:
            s.popleft()
            stack.pop()
        else:
            stack.append(s.popleft())        
        
    if len(stack) == 0:
        return 1
    else:
        return 0


s = 'baabaa'
solution(s)
````
