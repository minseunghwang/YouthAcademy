# Matplotlib에서 그래프를 그리는 2가지 방법

- plt.plot()
- add_subplot()

```python
# plt.plot() 방식

# 1) fig 객체 생성
fig = plt.figure()

# 2) 데이터 생성
# np.linspace(start, end, size)
x = np.linspace(0,2,100)

# 3) 데이터 전달
plt.plot(x,x**3, label='샘플')

# 4) 그래프 제목, 범례 등 지정
plt.title('그래프 예제')
plt.legend()

# 5) 그리드 표시
plt.grid()

# 6) 그래프 실행
plt.show()
```

```python
# add_subplot()

# 1) fig 객체 생성
fig = plt.figure(dpi=150)
# plt.show()

# 2) ax 변수 생성
# ax변수 = fig객체변수.add_subplot(행수,열수,순서)
ax = fig.add_subplot(1,1,1)
# ax

# 3) 데이터 생성
# np.linspace(start, end, size)
x = np.linspace(0,2,100)

# 4) 데이터 전달
# ax변수.plot(x데이터,y데이터, label=라벨명)
ax.plot(x,x**3,label='쌤플')

# 5) 그래프 제목
# plt.title('그래프 예제')
ax.set_title('그래프 제목')

# 6) 그리드 표시
# ax변수.grid()
ax.grid()

# 7) 범례 속성 생성
# ax변수.legend(속성=값)
# loc = 위치값, facecolor = 배경색상, borderaxespad = 테두리 굵기, edgecolor = 테두리 색상, fontsize = 글자크기
ax.legend(loc=1, facecolor='gray', borderaxespad=2, edgecolor='darkblue', fontsize=15)

# 8) 범례 표시
plt.show()
```
