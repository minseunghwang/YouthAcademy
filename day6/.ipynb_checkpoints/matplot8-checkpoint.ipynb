## Scatter Plot - 산점도

- plt.scatter(xData, yData, s=Size, c=colorName, marker=markerName)
- s는 사이즈
- value 표시 : plt.text(x좌표, y좌표, value)

```python
# 데이터 생성
names = ['a','b','c','d','e']
# values = np.random.randint(30,50,5)
values = np.random.randint(10,200,5)
print(values)
```

```python
# 그래프 생성
fig3 = plt.figure(dpi=100)
# fig3, plt.scatter(names, values, s=180, marker='+')
fig3, plt.scatter(names, values, s=values*70-50, c='blue', alpha=0.5)

# dof 라벨표시
# plt.text(xPos, yPos, value)

for i in range(0,len(names)):
    fig3, plt.text(names[i], values[i]*0.5, '<' + str(values[i]) + '>')
    
fig3, plt.title('산점도 그래프', fontsize=17, x=0.5, y=1.05)
fig3, plt.show()
```

## Pie Plot 
- plt.pie(values, labels=labelList, autopct = '%digit.pointNum%Unit')

```python
labels='Frogs', 'Hogs','Dogs', 'Logs'
values = [15, 30, 45, 10]

fig4 = plt.figure(dpi=100)
fig4, plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('파이그래프')
plt.show()
```

## Pie Plot - 슬라이스
- plt.pie(values, labels=labels, explode=explodeList, autopct = '%digit.pointNum%Unit', shadow=True, startangle=angleValue)
- explode : 슬라이싱되는 값

```python
# 데이터 생성
labels='Frogs', 'Hogs','Dogs', 'Logs'
sizes = [15, 30, 45, 10]
# 슬라이싱 리스트
explode = (0.05, 0.1, 0.15, 0.05)

fig6 = plt.figure(dpi=100)
fig6, plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=180)


plt.show()
```
