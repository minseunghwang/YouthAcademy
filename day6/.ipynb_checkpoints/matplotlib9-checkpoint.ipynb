# Matplotlib 시각화 - 데이터프레임 적용
- 데이터 시각화를 위한 파이썬 라이브러리
- https://matplotlib.org/

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.font_manager as fm

#  한글글꼴로 변경
plt.rcParams['font.size'] = 20.0
plt.rcParams['font.family'] = 'batang'
# plt.rcParams['font.family'] = 'Malgun Gothic'
print(plt.rcParams['font.family'])
print(plt.rcParams['font.size'])
```

## 데이터프레임 이용 -Line plot 1

- 데이터프레임 생성 : pd.DataFrame(데이터값, index=인덱스값, columns=컬럼값)
- 그래프 생성 : 데이터프레임변수.plot(kind=그래프종류)

```python
# 데이터 생성 => y
# np.random.randint(start, end, size)
data = np.random.randint(1,50,20).reshape(5,4)
data
```

```python
# 데이터 생성 => x => index
x = np.arange(0,50,10)

# 데이터 프레임 생성
df = pd.DataFrame(data, columns=['A','B','C','D'], index=x)
df
```

```python
# df.plot()
df.plot(kind='bar')
# df.plot(kind='barh')
```

### 데이터프레임의 선그래프

```python
fig = plt.figure(dpi=150)

# fig, df.plot()
# 선색상 모두 동일
# fig, df.plot(color='black')

# 선 속성 정의
fig, df['A'].plot(color='brown', style='--')
fig, df['B'].plot(color='red',  style='-*')
fig, df['C'].plot(color='blue',  style=':')
fig, df['D'].plot(color='lightgreen',  style='-.')
plt.show()

# 그래프에 전달 => 선 색상 모두 동일
# fig, df.plot(color='black')


# 신속성 각각 정의
# plt.grid()
# plt.legend(loc='upper right', fontsize=12, facecolor='lightgreen')
```

```python
# 별도
fig = plt.figure(dpi=150)

# 속성값 지정
fig, df.plot( color=['black','red','darkblue','green'],
            style={'A':'--','B':'-.','C':':','D':'--'},
            fontsize=11,
            grid=True,
            title='데이터프레임 직선 그래프',
            xticks=range(0,42,5), figsize=(15,8), xlim=(0,40)
            )
plt.show()
```

### 데이터프레임 그래프의 layout 속성을 이용하여 컬럼별 그래프 만들기

- 데이터프레임변수.plot(subplots=True, layout=(2,2))

```python
# 별도
fig = plt.figure(dpi=150)

# 2행 2열 배치되게 속성값 지ㅓㅈㅇ
fig, df.plot( color=['black','red','darkblue','green'],
            style={'A':'--','B':'-.','C':':','D':'--'},
            fontsize=11,
            grid=True,
            title='데이터프레임 직선 그래프',
            xticks=range(0,42,5), figsize=(15,8), xlim=(0,40),
            subplots=True,
            layout=(2,2)
            )
plt.show()
```

# add_subplot을 이용한 데이터프레임 그래프

- ax.변수.plot(데이터프레임)

```python
fig = plt.figure(dpi=120)
ax = fig.add_subplot(1,1,1)
# 데이터 프레임 데이터 전달
ax.plot(df)
# 범례 라밸을 데이터프레임.columns
ax.legend(df.columns, loc='upper right', fontsize=12,facecolor='lightgreen')
plt.grid()
plt.show()
```

# 데이터프레임의 인덱싱/슬라이싱 기능으로 그래프 만들기

```python
# 특정 컬럼만 그래프화
# 데이터프레임변수[컬럼명].plot()

df['A'].plot(figsize=[8,5], color='yellowgreen', style='--', title='A컬럼 직선 그래프')
```

```python
# 다중컬럼 그래프화
# 데이터프레임변수[[컬럼명1, 컬럼명2...]].plot(속성=값...)

df[['A','C']].plot(figsize=[8,5], color=['blue','red'], style=['--','-.'], title='A와 C컬럼 직선 그래프')
```

```python
# 특정행만 그래프화 (2행만 그래프화)

# df.loc[10].plot(figsize=[10,6])
df.iloc[1].plot(figsize=[10,6])

# 마지막행만 그래프화
df.iloc[-1].plot(figsize=[12,8], color='red', style='--', grid=True)
plt.legend(loc='upper center')
```

## 퀴즈 - 특정 데이터프레임 데이터를 출력하여 그래프화한다.

- 1) 짝수행만 그래프화
- 2) 1~3행중 A,B컬럼만 그래프화
- 3) B,C 컬럼을 제외하고 그래프화

```python
#1
df.iloc[range(0,len(df),2)].plot(figsize=[8,6], style=['--','-.','-*',':'], linewidth=3)
```

```python
#2
df.iloc[range(1,4)][['A','B']].plot(figsize=[8,6])
```

```python
#3
temp =[]
for i in df.columns:
    if i != 'B' and i != 'C':
        temp.append(i)
df[temp].plot(figsize=[8,6])
```
