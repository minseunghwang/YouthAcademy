# barh plot
- 가로막대그래프
- plt.barh(x,y,color=colorname, width=float)

```python
people = ('Tom','Dick','Harry','Slim','Jim','Kyle')
x_pos = np.arange(len(people))
print(x_pos) # [0 1 2 3 4 5]

# 이하 데이터값 -6개 : people 개수만큼
performance = np.random.rand(len(people))
performance
```

```python
# fig 생성
fig1 = plt.figure(dpi=120)
# 데이터 전달
# plt.barh(x_pos, performance, color='green')
plt.barh(x_pos, performance, color=['red','green','blue'])


# 눈금값 설정
plt.yticks(x_pos)
plt.xlabel('Performance')
plt.ylabel('people')
plt.title('Today\ man', x=0.5, y=1.05)

fig1, plt.show()
```

# barh 2 : add_subplot

```python
fig2 = plt.figure(frameon = True, dpi = 90)
ax = fig2.add_subplot(1,1,1)


# 그래프 생성
print(x_pos)
print(performance)
ax.barh(x_pos, performance, color='green', linestyle='solid')

# 눈금값 설정
ax.set_yticks(x_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Performence')
ax.set_title('Today\ Man')

plt.show()
```

## Scatter Plot - 산점도

- plt.scatter(xData, yData, s=Size, c=colorName, marker=markerName)
- s는 사이즈
- value 표시 : plt.text(x좌표, y좌표, value)

```python
# 데이터 생성
names = ['a','b','c','d','e']
# values = np.random.randint(30,50,5)
values = np.random.randint(10,200,5)
print(values)
```

```python
# 그래프생성 
fig3 = plt.figure(dpi=100)
# fig3, plt.scatter(names, values, s=50, c='blue', marker='+')
fig3, plt.scatter(names, values, s=values*70-50, c='blue', alpha=0.5)
# 라벨 표시 : plt.text(xPos, yPos, values)
# fig3, plt.text(names[0], values[0]+0.5, str(values[0]))
# fig3, plt.text(names[1], values[1]+0.5, str(values[1]))
# fig3, plt.text(names[2], values[2]+0.5, str(values[2]))
# fig3, plt.text(names[3], values[3]+0.5, str(values[3]))
# fig3, plt.text(names[4], values[4]+0.7, str(values[4]))

# for 문으로 변경하기 
for i in range(0, len(names)):
    fig3, plt.text(names[i], values[i]+ 0.5, '< ' + str(values[i]) + ' >')
fig3, plt.title('산점도 그래프', fontsize=17, x=0.5, y=1.05)   
fig3, plt.show()
```

# add_subplot() 방식으로 변경하기

```python
# 그래프생성 
fig3 = plt.figure(dpi=100)
axes = fig3.add_subplot(1,1,1)
axes.scatter(names, values, s=values*70-50, c='blue', alpha=0.5)

# for 문으로 변경하기 
for i in range(0, len(names)):
    axes.text(names[i], values[i]+ 0.5, '< ' + str(values[i]) + ' >')
    
axes.set_title('산점도 그래프', fontsize=17, x=0.5, y=1.05)   
fig3, plt.show()
```

### Scatter Plot - colorRandom, sizeRandom
- Dot의 크기와 색상 변화


```python
x = ['a','b','c','d','e','f', 'g']
y = np.random.rand(len(x))
dotcolor =  np.random.rand(len(x))
```

```python
fig, ax = plt.subplots(dpi=120)
ax.scatter(x, y, s=y*500+100, c=dotcolor, alpha=0.5)
# 눈금값 설정 
ax.set_xticklabels(x)
# x 축 이름 설정 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('색상과 크기가 변경되는 산점도 그래프')
plt.show()
```

## Pie Plot - 슬라이스
- plt.pie(values, labels=labels, explode=explodeList, autopct = '%digit.pointNum%Unit', shadow=True, startangle=angleValue)

```python
# 데이터 생성
labels='Frogs', 'Hogs','Dogs', 'Logs'
sizes = [15, 30, 45, 10]
# 슬라이싱 리스트
explode = (0.05, 0.1, 0.15, 0.05)

fig6 = plt.figure(dpi=100)
fig6, plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=180)


plt.show()
```

## Histogram plot
- figure 생성
- plt.hist(dataVales, bins=number)


```python
points = np.random.randn(10000)
print(points)
fig7 = plt.figure(dpi=100)
fig7, plt.hist(points, bins=20, color='hotpink')
plt.title('히스토그램')
plt.show()
```

## Histogram + Subplot
- figureVar, ax변수 = plt.subplots(row, col, sharey=True)
    - sharey=True : 간격 자동 지정
- ax변수[index].hist(dataVales, bins=number, color='colorName')

```python
x = np.random.randn(100000)
y = np.random.randn(100000)
fig8  = plt.figure(figsize=(12,12))
fig8 , axs = plt.subplots(1, 2, sharey=True)
fig8, axs[0].hist(x, bins=20, color='yellowgreen')
fig8, axs[1].hist(y, bins=30, color='brown')
plt.show()
```
