````python
import numpy as np
import pandas as pd
````

# 서로 다른 데이터프레임 join 하기

### 3개의 샘플 데이터프레임 생성
- 조건 : 컬럼명이 같아야 한다.

````python
df1 = pd.DataFrame(np.arange(1,17).reshape(4,4), columns=['A', 'B', 'C', 'D'])
df1
````

````python
df2 = pd.DataFrame(np.arange(2,18).reshape(4,4), columns=['A', 'B', 'C', 'D'])
df2
````

````python
df0 = pd.concat([df1,df2], ignore_index=True)
df0
# df00 = pd.concat([df0,df1])
# df00
````

# concat() 메소드 이용하기

- 새로운데이터프레임 = pd.concat([데이터프레임1, 데이터프레임2...])

````python
result_df = pd.concat([df1, df2])
result_df.loc[1]
result_df
````

#  append() 메소드 이용하기

- 데이터프레임 = 데이터프레임.append(딕셔너리, ignore_index=True)

````python
data = [[159,154,139,369],[488,693,777,530]]
result_df = result_df.append(data, ignore_index=True)
````

````python
dict1 = {'A':100, 'B':200, 'C':300, 'D':400}
df1 = df1.append(dict1, ignore_index=True)
df1
````

#  concat() 메소드 이용하기 2

- 가로로 붙이기(열 증가)

- 적용 데이터프레임의 행수가 다른 경우 누락값(NaN)발생

#### 결국 concat으로 행증가, 열증가 다 할 수 있네

````python
df3 = pd.DataFrame(np.arange(2,18).reshape(4,4), columns=['A', 'B', 'C', 'D'])
df3
````

````python
df53 = pd.concat([df1,df2], ignore_index=True, axis=1)
df53
````

# 이너조인(Inner Join)옵션 이용하기

- 공통컬럼만 골라 연결하기
```
    pd.concat([데이터프레임1, 데이터프레임2...], ignore_index=True, join='inner')
```
- 공통행만 골라 연결하기
```
    pd.concat([데이터프레임1, 데이터프레임2...], axis=1, join='inner')
```
- join='inner'

````python
df1 = pd.DataFrame(np.arange(1,17).reshape(4,4))
df2 = pd.DataFrame(np.arange(1,17).reshape(4,4))
df3 = pd.DataFrame(np.arange(1,17).reshape(4,4))
````

````python
df1.columns = ['A','B','C','D']
df2.columns = ['e','f','g','h']
df3.columns = ['A','o','C','p']
````

````python
# 3개의 데이터프레임을 inner join
result_df1 = pd.concat([df1,df2,df3],join='inner')
result_df1
result_df1.shape
````

````python
# 2개의 데이터프레임을 inner join
result_df2 = pd.concat([df1,df3], join='inner')
result_df2.shape
result_df2
````

````python
# 2개의 데이터프레임을 inner join
result_df3 = pd.concat([df1,df2], join='inner')
result_df3.shape
result_df3
````

````python
# 인덱스명 변경하기
df2.index = [2,3,4,5]
df3.index = [2,4,5,100]
````
