# 행 인덱싱

- 데이터프레임명[인덱스:인덱스+1] : 1개만 추출
- 데이터프레임명.loc[인덱스명]
- 데이터프레임명.iloc[위치인덱스]
- 데이터프레임명.head(위치인덱스)

```python
# 첫번째 1행 인덱싱하기 1 => 데이터 프레임
print(df.head(1)) # df.head(0) 하면 안댐
print('*'*50)

# 첫번째 1행 인덱싱하기 2 => 데이터 프레임
print(df[0:1])
print('*'*50)

# 첫번째 1행 인덱싱하기 3 => 시리즈
print(df.loc[0])
print('*'*50)

# 첫번째 1행 인덱싱하기 4 => 시리즈
print(df.iloc[0])
print('*'*50)


print('예시@@@@@@@@@@@@@@@@@@@@@@@@@@@@')

print(df.loc[3:5])
print('*'*50)
print(df.iloc[3:7])
```

```python
# 마지막 행 출력
print(df.tail(1)) # df.tail(0) 하면 안댐

# 전체 행수
print(f'전체 행수 : {df.shape[0]}')
print(f'전체 행수 : {len(df)}')
```

```python
# 다같은거
print(df[len(df)-1:])
print('*'*50)
print(df.loc[len(df)-1])
print('*'*50)
print(df.iloc[-1])
print(df.iloc[len(df)-1])
```

## 다중 행 추출

- 겹대괄호 이용[[]]
- 데이터프레임명.loc[[인덱스명1, 인덱스명2..]]
- 데이터프레임명.iloc[[위치인덱스1, 위치인덱스2..]]

```python
# 1,3,5,7,9 행을 출력하여라
print(df.loc[[0,2,4,6,8]])
print(df.iloc[[0,2,4,6,8]])
```

```python
# 0~50 인덱스중에서 짝수만 출력
print(df.loc[0:50:2])
df.loc[np.arange(1,50,2)]
```
