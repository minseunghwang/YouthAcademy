```python
import numpy as np
import pandas as pd
```

# 행 추가하기

- 데이터프레임.loc[인덱스명]=[값1, 값2...]
- 데이터프레임 = 데이터프레임.append([[값1, 값2 ...]], ignore_index=True)
        - ignore_index=True 옵션값 사용시 인덱스명이 숫자로 자동 리셋
- 데이터프레임 = 데이터프레임.append([값1, 값2...])
        - 값이 1컬럼만 삽압된다. 나머지는 NaN으로 삽입된다.

```python
# 샘플 데이터프레이를 넘파이 이용해서 만들기

n1 = np.arange(1,21).reshape(4,5)
n1
```

```python
n1.ndim # 차원 확인
```

```python
df1 = pd.DataFrame(n1, index=['one','two','three','four'])
df1
```

```python
# 새로운 행 추가하기
df1.loc['five'] = 1
df1.loc['six'] = [21,22,23,24,25]
df1
```

```python
# append() 메소드를 이용하여 새로운 행 추가하기
df1 = df1.append([[100,90,80,70,60]], ignore_index=True)
df1
```

```python
df1 = df1.append([100,99,98,97,96])
df1
```

```python
df1 = df1.append([[100,90,80,70,60], [200,190,180,170,160]])
df1
```
