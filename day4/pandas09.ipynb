````python
#1
df_movie = pd.read_csv('../data/movies.dat', sep='::', header=None, index_col=0)
df_movie
````

````python
#2 
df_movie.columns=['Title', 'Kind']
df_movie.index=range(0,len(df_movie))
df_movie
````

````python
# in연산자 : item/문자 in 집합형자료(list/시리즈/튜플...)/문자열

print('a' in 'apart') # True
print('b' in 'apart') # False
print(100 in [1,90,100]) # True

#3
cnt = 0
for i in df_movie['Title']:
    if i[-5:-1] == '1995':
        cnt +=1
print(cnt)
##############################
for i in range(len(df_movie)):
    '1995' in df_movie['Title'].loc[i]
    
df_movie.loc[0,'Title']
df_movie.loc[0][0]
df_movie.iloc[0,0]
````

````python
# 4
def movieyear(n):
    cnt = 0
    for i in df_movie['Title']:
        if i[-5:-1] == str(n):
            cnt +=1
    return cnt
    
movieyear(1995)
````
