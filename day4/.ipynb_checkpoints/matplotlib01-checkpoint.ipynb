### matplotlib

- 데이터 시각화를 위한 파이썬 라이브러리

```python
# 패키지 임포트

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Figure 객체 생성하기

- Figure : 도식, 그래프가 삽입되는 객체
- var = plt.figure()

```python
fig = plt.figure()
type(fig)
```

# plot 객체 생성하기

- 피규어변수, plt.plot(xData, yData, label=labelName) : figure에 plot 생성하기
- plt.xlabel(X축)
- plt.ylabel(Y축)
- plt.title(제목)
- plt.legend() : 범례 표시
- plt.show() : plot 다시 그리기

```python
# x,y 데이터 생성 - 리스트로 생성시 개수가 같아야한다.
x = np.arange(0,5)
y = [10,25,30,8,30]
plt.plot(x,y,label='Test')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Main Title')
plt.legend()
plt.show()
```

# 다중 그래프
# 첫번째 그래프
- 피규어 변수, plt.plot(xData1, yData1, label='라벨명1')
# 두번째 그래프
- 피규어 변수, plt.plot(xData2, yData2, label='라벨명2')

# 옵션
- plt.xlabel(X축)
- plt.ylabel(Y축)
- plt.title(제목)
- plt.legend() : 범례
- plt.show() : plot다시그리기


```python
x = np.arange(0,5)
y = [10,25,30,8,30]
# 첫번째 그래프
fig, plt.plot(x,label='graph1', color='#FF0000')

# 두번째 그래프
fig, plt.plot(y,label='graph2', color='#00FF00')

# 세번째 그래프
fig, plt.plot(x**2,label='graph3', color='#0000FF')

# 16진수 색상값이란?
#000000 ~ #FFFFFF


plt.plot(x,y,label='Test')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Main Title')
plt.legend()
plt.show()
```


### 색상 사용하기
plt.plot(x,y, c='아래의문자중하나') plt.plot(x,y, color='16진수색상값')

``'b'``          blue
``'g'``          green
``'r'``          red
``'c'``          cyan
``'m'``          magenta
``'y'``          yellow
``'k'``          black
``'w'``          white
선의 스타일 지정
plt.plot(x,y, linestyle=선값)

``'-'``          solid line style
``'--'``         dashed line style
``'-.'``         dash-dot line style
``':'``          dotted line style

```python
x = np.arange(0,12,0.1)
y = np.sin(x)

fig4 = plt.figure()

fig4, plt.plot(x,y, label='sin1', c='m', linestyle='--')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Math Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.arange(0, 12, 0.1)
y = np.sin(x)


fig4 = plt.figure()
fig4, plt.plot(x, y, label='sin1', c='m', linestyle='--')
fig4, plt.plot(x, -y, label='sin2', linestyle='-.')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Math Graph')
plt.legend()
# plt.grid()
plt.show()
```

### 마커 사용하기
https://www.w3schools.com/colors/colors_names.asp

    plt.plot(x, y,  marker=마커값, 
                    markerfacecolor=16진수색상값/영문색상명, 
                    markersize=floatValue)


``'.'``          point marker
``','``          pixel marker
``'o'``          circle marker
``'v'``          triangle_down marker
``'^'``          triangle_up marker
``'<'``          triangle_left marker
``'>'``          triangle_right marker
``'1'``          tri_down marker
``'2'``          tri_up marker
``'3'``          tri_left marker
``'4'``          tri_right marker
``'s'``          square marker
``'p'``          pentagon marker
``'*'``          star marker
``'h'``          hexagon1 marker
``'H'``          hexagon2 marker
``'+'``          plus marker
``'x'``          x marker
``'D'``          diamond marker
``'d'``          thin_diamond marker
``'|'``          vline marker
``'_'``          hline marker

```python
x = np.arange(-10, 10)
y = x**2

fig5 = plt.figure()
fig5, plt.plot(x, y, label='markerStyle1', marker='D', markerfacecolor='HotPink',  
                      markersize=7.0)
fig5, plt.plot(x, -y, label='markerStyle2', marker='X', markerfacecolor='yellowgreen',  
                      markersize=5)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Math Graph')
plt.legend()
# plt.grid()
plt.show()
```
