- inplace = True : 원본에 적용

```python
df = pd.DataFrame(np.random.randn(6,4), columns=['C','A','D','B'])

df.index = [2,5,1,0,3,4]
df
```

```python
# 복사본 만들기
df2 = df.copy()

# 행 방향으로 오름차순 정렬
df2.sort_index(axis=0)   # 머냐이거 데이터정렬이아니고 index정렬이네
```

```python
# 열 방향으로 오름차순 정렬
df2.sort_index(axis=1)    # 얘도 A B C D 만 순서대로됫는디 ?
```

# groupby

- 데이터프레임.groupby([컬럼명]).집계함수
- 집계함수 : min(), max(), mean(), sum(), median(), std()

```python
# child / man/ woman 의 평균 나이와 생존률 구하기
result.groupby(['who']).mean()[['age','survived']]
```

```python
# 생존자와 사망자 중에서 가장 나이가 많은사람 
result.groupby(['who']).max()['age']
```

```python
# 각 등급별 생존률
result.groupby(['class']).mean()['survived']
```

```python
result.groupby(['embarked']).mean()[['fare','age']]
```

```python
# 타이타닉 csv 파일 이용하기
# PassengerId를 인덱스 명으로 이용
train_df = pd.read_csv('../data/train.csv', index_col='PassengerId')
train_df
```

```python
# 1등칸 객실 승객중에서 10명의 데이터 출력
# 1등칸 명부
train_df[train_df['Pclass']==1][['Name','Age','Cabin']].head(10)
```

```python
# Cabin의 누락값 개수는?
np.count_nonzero(train_df['Cabin'].isnull())
# train_df['Cabin'].isnull().value_counts() # 누락된거 안된거 둘다보기
# train_df['Cabin'].count() # 이건 누락되지않은값 
```

# 퀴즈
- 1등칸 객실 승객의 생존 리스트 출력하기
- 1등칸의 생존률은?
- 여자 생존자수 ? 남자 생존자수 ?
- 남자와 여자의 생존률 구하기
- Age 컬럼값에서 결측값(NaN) 총 갯수는?
- Age 컬럼값의 결측값을 Age의 평균값으로 교체하기
- 'Embarked'컬럼에서 아래와 같은 데이타값 교체하기
    - C -> France
    - Q -> New Zealand
    - S -> England
- 'Embarked'컬럼을 groupby하여 평균 나이 출력하기
- 가족을 동반하지 않고 승선한 승객의 남녀 비율 출력하기
- 생존자와 사망자의 데이타프레임을 생성하고 csv 파일로 각각 저장하기

```python
# 1
a = train_df[train_df['Survived']==1]
a[a['Pclass']==1]

# 2
train_df[train_df['Pclass']==1].mean()['Survived']

# 3
print('female생존자수 : ', a['Sex'].value_counts()[0]) # female 생존자수
print('male생존자수 : ', a['Sex'].value_counts()[1]) # male 생존자수

# 4
train_df[train_df['Survived']==0]['Sex'].value_counts()[0] # male 사망자수 : 468
train_df[train_df['Survived']==0]['Sex'].value_counts()[1] # female 사망자수 : 81
male_per = a['Sex'].value_counts()[0] / train_df['Sex'].value_counts()[0]
female_per = a['Sex'].value_counts()[1] / train_df['Sex'].value_counts()[1]
print(f'male 생존률 : {male_per}') # male 탑승자 수 :577
print(f'female 생존률 : {female_per}') # female 탑승자 수 :314

# 5. Age 컬럼값에서 결측값(NaN) 총 갯수는?
np.count_nonzero(train_df['Age'].isnull())

# 6. Age 컬럼값의 결측값을 Age의 평균값으로 교체하기
train_df['Age'].fillna(train_df['Age'].mean())

# 7. 'Embarked'컬럼에서 아래와 같은 데이타값 교체하기
#     - C -> France
#     - Q -> New Zealand
#     - S -> England
train_df['Embarked'].replace('C', 'France', inplace=True)
train_df['Embarked'].replace('Q', 'New Zealand', inplace=True)
train_df['Embarked'].replace('S', 'England', inplace=True)
train_df['Embarked']

# 8. 'Embarked'컬럼을 groupby하여 평균 나이 출력하기
train_df.groupby(['Embarked']).mean()['Age']

# 9. 가족을 동반하지 않고 승선한 승객의 남녀 비율 출력하기
nofam=train_df[train_df['SibSp']==0]
nofamsum = nofam[train_df['Parch']==0]['Sex'].value_counts().sum()
nofam[train_df['Parch']==0]['Sex'].value_counts()/nofamsum

# 10. 생존자와 사망자의 데이타프레임을 생성하고 csv 파일로 각각 저장하기
survivor = train_df[train_df['Survived']==1] # 생존자
dead = train_df[train_df['Survived']==0] # 사망자

survivor.to_csv('../output/train_survivor.csv')
dead.to_csv('../output/train_dead.csv')
```
